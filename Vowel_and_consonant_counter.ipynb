```python
# Vowel and Consonant Counter
# Problem Statement
# Develop a Python program to count vowels and consonants in a string.
# Requirements
# Accept a string from the user.
# Count:
# Number of vowels
# Number of consonants
# Number of digits
# Number of special characters
# Display the results.
# Ignore spaces.



string = input("Enter a string: ")
vowels = 0
consonants = 0
digits = 0
special_characters = 0

for char in string:
    if char.lower() in "aeiou":
        vowels += 1
    elif char.isalpha():
        consonants += 1
    elif char.isdigit():
        digits += 1
    else:
        special_characters += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special Characters:", special_characters)
```
