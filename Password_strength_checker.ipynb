Problem Statement
Determine whether a password is weak, medium, or strong.
Requirements
Check:
• Length ≥ 8
• Contains uppercase letter
• Contains lowercase letter
• Contains digit
• Contains special characters 
Display:
• Weak
• Medium
• Strong

```python
password = input("Enter a password to check its strength: ")
length = len(password)
count=0
for char in password:
    if length>=0:
        count+=1
    elif char.isupper():
        count+=1
    elif char.islower():
        count+=1
    elif char.isdigit():
        count+=1
    elif not char.isalnum():
        count+=1

if count >= 5:
    print("Password strength: Strong")
elif length >= 2 and count <5 :
    print("Password strength: Medium")
else:
    print("Password strength: Weak")
```
